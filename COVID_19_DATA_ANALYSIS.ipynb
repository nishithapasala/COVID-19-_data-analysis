```python
# removing unneccessary columns or droping them
covid_df=pd.read_csv('/content/drive/MyDrive/DATA ANALYSIS/COVID-19 DATA ANALYSIS/covid_19_india.csv')
print(covid_df)
'''
covid_df.drop(['Sno', 'Time', 'ConfirmedIndianNational', 'ConfirmedForeignNational'], inplace = True, axis = 1)'''

```

```python
covid_df.head()
```

```python
#change format of date column
covid_df['Date']= pd.to_datetime(covid_df['Date'], format ='%Y-%m-%d')
covid_df.head(6)
```

```python
# Active cases

covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.head()
```

```python
#pivot table using pandas
statewise = pd.pivot_table(covid_df, values = ["Confirmed" , "Deaths" , "Cured"], index ="State/UnionTerritory" , aggfunc = max)


statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
covid_df.head(10)
```

```python
top_10_active_cases = (
    covid_df.groupby('State/UnionTerritory')[['Active_Cases', 'Date']]
    .max()
    .reset_index()
    .sort_values(by='Active_Cases', ascending=False)
    .head(10)
)
```

```python

# Set the figure size
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed

# Create the bar plot
x = sb.barplot(
    data=top_10_active_cases,
    y="Active_Cases",  # Correct column name
    x="State/UnionTerritory",  # Ensure column name matches
    linewidth=2,
    edgecolor='red'
)

# Add title and axis labels
x.set_title("Top 10 States with Highest Active Cases", fontsize=16)
x.set_xlabel("State/Union Territory", fontsize=12)
x.set_ylabel("Active Cases", fontsize=12)
plt.xticks(rotation=45, fontsize=10)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```

```python
# Top 10 states with highest deaths

top_10_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = ['Deaths'],ascending = False).reset_index()
fig = plt.figure(figsize=(18,10))
plt.title("Top 10 Deaths", size = 25)

x = sb.barplot(
    data=top_10_deaths.iloc[:12],
    y="Deaths",  # Correct column name
    x="State/UnionTerritory",  # Ensure column name matches
    linewidth=2,
    edgecolor='red'
)

plt.xlabel("states")
plt.ylabel("Deaths")
plt.show()
```

```python
# Growth trend
# Set the figure size
fig = plt.figure(figsize=(12,5))

# Create lineplot
x = sb.lineplot(
    data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharastra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])],
    y="Active_Cases",  # Correct column name
    x="Date",  # Correct column name
    hue="State/UnionTerritory"  # Group by state
)

# Set the title
x.set_title("Top 5 Affected States in India", size=19)

# Adjust x-axis label rotation for better readability
plt.xticks(rotation=45, fontsize=10)

# Show the plot
plt.show()
```

```python
#remove unwanted columns
vaccination = vaccine_df.drop(columns = ['Sputnik V (Doses Administered)' , 'AEFI', '18-44 Years (Doses Administered)', '45-60 Years (Doses Administered)', '60+ Years (Doses Administered)'],axis = 1)
```

```python

# Calculate the total male and female vaccinations
male = vaccination["Male(Individuals Vaccinated)"].sum()
female = vaccination["Female(Individuals Vaccinated)"].sum()

# Create the pie chart
fig = pxi.pie(
    names=["Male", "Female"],  # Labels for the slices
    values=[male, female],     # Values for the slices
    title="Male and Female Vaccination"
)

# Show the pie chart
fig.show()
```

```python
# remove rows where state = India

vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns = {"Total Individuals Vaccinated": "Total"}, inplace = True)
vaccine.head()
```

```python
# most vaccinated State

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total' , ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize = (10,5))
plt.title("Top 5 vaccinated States in India", size = 20)
x = sb.barplot(data = max_vac.iloc[:10],y = max_vac.Total, x = max_vac.index, linewidth=2, edgecolor='green')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
